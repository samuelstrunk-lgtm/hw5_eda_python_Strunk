```python
# Now, I first started by simply importing the main libraries that were needed for this assignment
# I just used pandas in order to load and work with the dataset
import pandas as pd

# I used numpy because it can be very useful for a few small numeric tasks
import numpy as np

# Then I imported matplotlib because it is the base plotting library for this assignment
import matplotlib.pyplot as plt

# I also imported seaborn because it makes the graphs look a lot cleaner
# It also helps a lot with the faceted plots
import seaborn as sns

# Then I used this function because it actually makes plots show up directly inside the Jupyter notebook
%matplotlib inline

# Then I used this function because it actually gives the charts a cleaner default appearance
sns.set_theme(style="whitegrid")
```

```python
# Now, here I just simply told python the exact file name of the dataset
DATA_PATH = "kc_house_data_original.csv"

# Then this function reads the CSV file into a pandas dataframe
df = pd.read_csv(DATA_PATH)

# Then I printed the shape so that I can actually see how many rows and columns are loaded
print("Dataset shape:", df.shape)

# Then I simply displayed the first 5 rows just to confirm that the data loaded correctly
df.head()
```

```python
# Now, this just lets me see all of the column names that are in the dataset
# I actually found this very helpful because I wanted to match the assignment variables perfectly
print("Columns in the dataset:")
print(df.columns.tolist())

# Now, this just basically shows the data types and whether or not there are any missing values
print("\nDataFrame info:")
df.info()
```

```python
# Now, this just gives me summary statistics for the numeric columns
# This just helped me actually understand the center, spread , and range of the data itself
df.describe()
```

```python
# Now, the data column is currently stored like text, not as a real datetime value
# So basically, I need to actually convert if before I can group the sales by month
df["date"] = pd.to_datetime(df["date"], format="%Y%m%dT%H%M%S")

# Now, I simply created a new column  that actually keeps only the month part of the sale date
df["sale_month"] = df["date"].dt.to_period("M").astype(str)

# Then, I also created a year/month datetime version for just cleaner sorting and plotting
df["sale_month_dt"] = df["date"].dt.to_period("M").dt.to_timestamp()

# Then, I just ran this function in order to check the result to make sure the conversion actually worked
df[["date", "sale_month", "sale_month_dt"]].head()
```

```python
# So, this just simply shows how many missing values appear in each column
# I just did this because missing values can affect the plots
missing_values = df.isnull().sum()

# Then, I used this function to display the columns that actually have missing values
missing_values[missing_values > 0]
```

```python
# Now, for the scatter plot, I am going to compare the living area to price
# Then, I will use grade as the third variable by just mapping it to color
plt.figure(figsize=(10, 6))

# Now, seaborn scatterplot actually makes this easy and it looks very clean
# I even went as far as to use a little transparency because there are a lot of points
sns.scatterplot(
    data=df,
    x="sqft_living",
    y="price",
    hue="grade",
    alpha=0.6
)

# Now, I just added a title, and axis labels.
# I did this just to make the graph easier to read
plt.title("House Price vs Living Space, Colored by Grade")
plt.xlabel("Square Feet of Living Space")
plt.ylabel("Price")

# Then this just simply helps keep the legend from covering too much of the graph
plt.legend(title="Grade", bbox_to_anchor=(1.05, 1), loc="upper left")

# Then I actually used a tight layout just to simply help the chart fit better inside the actually notebook
plt.tight_layout()

# Then I used this function to display the graph
plt.show()
```

```python
# This scatter plot just show a positive relationship between both the living space and the house price.
# Now, in general, homes with more square footage tend to sell for a lot more. 
# The color mapping also suggests that the higher grade homes are often associated with higher prices.
# This actually makes total sense simply because grade reflects both construction and design quality.
```

```python
# Now, for the faceted histograms, I used waterfront because it only has two values
# This just simply makes the comparison easier to read
g = sns.FacetGrid(df, col="waterfront", height=5, aspect=1.2)

# Then I plotted a histogram of price inside each facet
# So basically, this just allows us to compare the price distribution for waterfront vs non waterfront homes
g.map_dataframe(sns.histplot, x="price", bins=30)

# Then I just added titles and labels to make the output look more complete
g.set_axis_labels("Price", "Count")
g.set_titles("Waterfront = {col_name}")

# Then this just adds an overall title for the faceted charts 
g.fig.suptitle("Distribution of House Prices by Waterfront Status", y=1.05)

# Then I used a tight layout to help everything fit verty neatly
plt.tight_layout()

# Then I used this function to show the faceted histograms
plt.show()
```

```python
# Now, these faceted histograms actually show that waterfront homes tend to have a different price distribution than non waterfront homes.
# Waterfront homes are are less commmon, but they actually appear more oftenj at the higher end of the price range.
# Now, non waterfront homes are much more common and are actually concentrated more heavily especially in the lower and middle price ranges. 
```

```python
# Next, I needed to count how many homes that are actually sold in each month
# Then I simply grouped by the month column that was created earlier
monthly_sales = (
    df.groupby("sale_month_dt")
      .size()
      .reset_index(name="homes_sold")
)

# Then I actually displayed the small montly table just to see what it looks like
monthly_sales
```

```python
# Next, I made the time series plot
# I am simply plotting the number of homes sold by month
plt.figure(figsize=(10, 6))

sns.lineplot(
    data=monthly_sales,
    x="sale_month_dt",
    y="homes_sold",
    marker="o"
)

# Next, I just simply added labels and a title to make the chart easier to read
plt.title("Number of Homes Sold by Month")
plt.xlabel("Month")
plt.ylabel("Number of Homes Sold")

# Then I just rotated the x-axis labels just to make the months easier to read
plt.xticks(rotation=45)

# I also used a tight layout in order to help avoid any text overlap
plt.tight_layout()

# Then I used this function in ofer to show the time series chart
plt.show()
```

```python
# Now, this time series plot actually shows how the number of homes sales changes from month to month.
# The plot actually helps reveal whether sales activity stayed fairly steady or whether there were stronger and weaker months.
# Now, this is actually very useful simply because it adds a time based view of the housing market rather than looking only at price realtionships. 
```

```python
# Now, for the extra plot, I decided to use a boxplot of price by bedrooms
# This right here actually helps show how the price distribution changes as the number of bedrooms changes
plt.figure(figsize=(10, 6))

sns.boxplot(
    data=df,
    x="bedrooms",
    y="price"
)

# Now, since a few extreme values can actually stretch the chart, I just limited the y-axis a little bit
# I just did this so that the main pattern is easier to see
plt.ylim(0, df["price"].quantile(0.95))

# Then I just added a title and labels
plt.title("House Price Distribution by Number of Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Price")

# I also used a tight layout to help the notebook look cleaner
plt.tight_layout()

# Then I used this function to display the chart
plt.show()
```

```python
# Now, this boxplot actually suggests that house price generally increases as the number of bedrooms increases, although the relationship is not perfectly consistent.
# Now, some bedrooms categories actually have a wide spread, which just basically means that homes with the same number of bedrooms can still vary a lot in price.
# In the end, this just tells us that bedrooms matter, but they are not the only facter affecting the price.  
```
